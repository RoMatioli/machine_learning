# K Means Clustering com Python

## Método usado

K Means Clustering é um algoritmo de aprendizagem sem supervisão que tenta agrupar dados com base em sua similaridade. A aprendizagem não supervisionada significa que não há resultados a serem previstos, e o algoritmo apenas tenta encontrar padrões nos dados. No K means clustering temos a especificar o número de clusters nos quais os dados devem ser agrupados. O algoritmo atribui aleatoriamente cada observação a um cluster e encontra o centróide de cada cluster. Então, o algoritmo itera através de duas etapas:
Reatribui pontos ao cluster cujo centroide é o mais próximo e calcula o novo centróide de cada cluster. Estes dois passos são repetidos até que a variação dentro do cluster não possa ser mais reduzida. A variação dentro do cluster é calculada como a soma da distância euclidiana entre os pontos de dados e seus respectivos centroides do cluster.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Criando dados

```python
# Criando os dados sintéticos usando make_blobs
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

# Visualizando os dados

```python
# Visualizando os dados gerados
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.title('Dados Originais')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Criando os clusters

```python
# Aplicando o algoritmo K-Means para criar os clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

O atributo cluster_centers_ retorna esses centróides finais após a execução do algoritmo. Geralmente, é útil para análise e visualização dos clusters formados. Por exemplo, você pode querer plotar os centróides junto com os pontos de dados para entender melhor a estrutura dos clusters.

```python
# Obtendo os centróides finais dos clusters
centroids = kmeans.cluster_centers_
print("Centróides finais:")
print(centroids)
```

O atributo labels_ retorna esses rótulos de cluster após a execução do algoritmo. Cada elemento no array labels_ corresponde ao rótulo do cluster para o ponto de dados correspondente no conjunto de dados.

```python
# Obtendo os rótulos de cluster atribuídos a cada ponto de dados
labels = kmeans.labels_
print("\nRótulos de cluster:")
print(labels)
```

```python
# Visualizando os clusters formados pelo K-Means e comparando com os dados originais
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('Clusters com K-Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=labels, cmap='rainbow')
ax1.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='black', label='Centróides')
ax1.legend()
ax2.set_title("Dados Originais")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')
plt.show()
```
